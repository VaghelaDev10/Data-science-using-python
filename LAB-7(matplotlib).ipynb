```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'b*-',linewidth=0.25)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.subplot(2,2,2)
plt.plot(x,y,'bs-',linewidth=0.25)
plt.subplot(2,2,3)
plt.plot(x,y,'bo-',linewidth=0.25)
plt.subplot(2,2,4)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'bs-',label='python')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.xlim(0,10)
plt.ylim(20,-10)
plt.title('This is Title')
plt.legend()


plt.subplot(1,2,2)
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
import numpy as np
import pandas as pd

np.random.seed(1)
df=pd.DataFrame(np.random.randint(1,100,(3,3)),index=['dev','Teerth','Fenil'],columns=['Python','Data Science','DBMS'])
df
```

```python
plt.plot(df.index,df.Python,'r-.',label='Python',lw='0.25',marker='x')#'rgbcymk' is the colour red
plt.title('Python Marks')

plt.plot(df.index,df.DBMS,'b--',label='DBMS')
plt.title('Student Marks')
plt.xlabel('Names')
plt.ylabel('Marks')

plt.plot(df.index,df['Data Science'],'g-*',label='Data Science')

plt.legend(loc=2)#loc=1,2,3,4 for position of legends

plt.ylim(-20,100)

## 1. solid(-) 2. dashed(--) 3. dashdot(-.) 4. dotted(:)
```

```python
plt.subplot(2,2,1)

plt.plot(df.index,df.Python,'r-.',label='Python',lw='0.75',marker='s')#'rgbcymk' is the colour red
plt.title('Python marks')
plt.subplot(2,2,2)
plt.plot(df.index,df.DBMS,'b--',label='DBMS',lw='0.25',marker='o')

plt.subplot(2,2,3)
plt.plot(df.index,df['Data Science'],'g-*',label='Data Science',marker='d')
plt.title('Data scidence marks')
plt.legend(loc=2)#loc=1,2,3,4 for position of legends

plt.ylim(-20,100)

plt.subplot(2,2,4)

plt.plot(df.index,df.Python,'r-.',label='Python',lw='1.25',marker='^')#'rgbcymk' is the colour red

plt.plot(df.index,df.DBMS,'b--',label='DBMS',lw='1.25',marker='^')

plt.plot(df.index,df['Data Science'],'g-*',label='Data Science',lw='1.25',marker='^')

plt.legend(loc=2)#loc=1,2,3,4 for position of legends

plt.ylim(-20,100)

plt.tight_layout()

## 1. solid(-) 2. dashed(--) 3. dashdot(-.) 4. dotted(:)
```

```python
x=np.linspace(0,6,11)
y=x**2

fig=plt.figure()

axes=fig.add_axes([0.01,0.01,0.5,0.5])
axes.plot(x,y,'b')
axes.set_xlabel('Set x label')
axes.set_ylabel('Set y label')
axes.set_title('Set title')
```

```python
x=np.linspace(0,6,11)
y=x**2

fig=plt.figure()

axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.6,0.2,0.2])

axes1.plot(x,y,'b--')
axes1.set_xlabel('Set x label')
axes1.set_ylabel('Set y label')
axes1.set_title('Set title')

axes2.plot(y,x,'r*-')
axes2.set_xlabel('Set x label')
axes2.set_ylabel('Set y label')
axes2.set_title('Set title')
```
